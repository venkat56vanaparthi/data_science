```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
number_df = DataFrame(np.arange(0, 90, 3).reshape(10, 3), index = ['row1','row2', 'row3', 'row4', 'row5', 'row6', 'row7', 'row8', 'row9', 'row10'], 
                      columns=['column1', 'column2', 'column3'])
number_df
```

```python
#simple index
number_df.iloc[0, 1]
```

```python
number_df.iloc[0, 0] = 20
number_df
```

```python
number_df.iloc[[1, 2, 3], [1, 2]]
```

```python
#boolean indexing
#comparision operators(< <= > >= = == !=) and masking

mask = number_df > 30
mask
```

```python
number_df[mask]
```

```python
number_df[number_df>30] = 0
number_df
```

```python
number_df.iloc[2:6, 1:3]
```
