```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
number_df = DataFrame(np.arange(0, 90, 3).reshape(10, 3), index = ['row1','row2', 'row3', 'row4', 'row5', 'row6', 'row7', 'row8', 'row9', 'row10'], 
                      columns=['column1', 'column2', 'column3'])
number_df
```

```python
#simple index
number_df.iloc[0, 1]
```

```python
number_df.iloc[0, 0] = 20
number_df
```

```python
number_df.iloc[[1, 2, 3], [1, 2]]
```

```python
#boolean indexing
#comparision operators(< <= > >= = == !=) and masking

mask = number_df > 30
mask
```

```python
number_df[mask]
```

```python
number_df[number_df>30] = 0
number_df
```

```python
#Missing values
#filling missing values using fillna(), replace() and interpolate()

data = {'name':['steve', 'john', 'richard', 'sarah', 'randy', 'micheal', 'julie'],
        'age':[20, 22, 20, 21, 24, 23, 22],
        'gender':['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female'],
        'rank':[2, 1, 4, 5, 3, 7, 6]}
ranking_df = DataFrame(data)
ranking_df.iloc[2:5, 1] = np.nan
ranking_df.iloc[3:6, 3] = np.nan
ranking_df.iloc[3, :] = np.nan

ranking_df
```

```python
ranking_df.isnull()
```

```python
ranking_df.notnull()
```

```python
bool_series = pd.isnull(ranking_df['age'])
ranking_df[bool_series]
```

```python
#filling with single value
ranking_df.fillna(0)
```

```python
ranking_df.fillna(method='pad')
```

```python
ranking_df.fillna(method='bfill')
```

```python
ranking_df.interpolate(method='linear')
```

```python
ranking_df.dropna()
```

```python
ranking_df.dropna(how='all')
```

```python
ranking_df.dropna(axis=1)
```

```python
ranking_df.dropna(axis=0)
```
